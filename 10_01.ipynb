```python
# 실습1. RGB 히스토그램 그리기
# 1.임의의 이미지를 선택하고
# 2.pyplot을 이용해서 RGB히스토그램을 한번에 그려보자
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("images/tea.jpg")

hist_1 = cv2.calcHist([img], [0], None, [256], [0,256])
hist_2 = cv2.calcHist([img], [1], None, [256], [0,256])
hist_3 = cv2.calcHist([img], [2], None, [256], [0,256])

plt.plot(hist_1, color = "b")
plt.plot(hist_2, color = "g")
plt.plot(hist_3, color = "r")

plt.title("histogram")
plt.xlabel("Pixel value")
plt.ylabel("Freqeuncy")
plt.show()
```
