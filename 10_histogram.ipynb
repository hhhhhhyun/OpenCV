```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("images/tea.jpg")
# channel : 0 - Blue, 1 - Green, 2 - Red
hist = cv2.calcHist([img], [1], None, [256], [0,256])

plt.plot(hist, color = "b")
plt.title("Histogram")
plt.xlabel("Pixel value")
plt.ylabel("Freqeuncy")
plt.show()
```
